Que:use the iris dataset from seaborn  to plot a pair plot . use the hue parameter for the "species" column of the iris dataset

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

load the fmri dataset using the load dataset fun of seaborn. plot a line plot using x= Timepoint and y= signal for different events and regions

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.relplot(x='timepoint', y='signal', hue= 'region', style='Event',
palette='ch:rot=-.25, hue=1,light=.75',
 markers=True, Kind='line',data=fmri)
```

Que: use the diamonds dataet from seaborn. to plot a histogram of seaborn. use the hue parameter for the cut column

```python
import seaborn as sns
```

```python
diamonds=sns.load_dataset('diamonds')
```

```python
df=diamonds
```

```python
df
```

```python
sns.lmplot(y="carat",x="price",hue="cut", data=df)
```
